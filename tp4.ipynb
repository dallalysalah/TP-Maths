```python
import numpy as np

def coin_nord_ouest(offre, demande):
    m = len(offre)
    n = len(demande)
    allocation = np.zeros((m, n))

    i = 0
    j = 0

    while i < m and j < n:
        alloc = min(offre[i], demande[j])
        allocation[i, j] = alloc
        offre[i] = offre[i] - alloc
        demande[j] = demande[j] - alloc

        if offre[i] == 0:
            i = i + 1
        elif demande[j] == 0:
            j = j + 1

    return allocation


def calculer_cout_total(allocation, couts):
    return np.sum(allocation * couts)


def afficher_solution(allocation, couts):
    m, n = allocation.shape

    print("\nSolution:")
    for i in range(m):
        for j in range(n):
            if allocation[i, j] > 0:
                print(f"x[{i+1},{j+1}] = {int(allocation[i, j])}")

    cout_total = calculer_cout_total(allocation, couts)
    print(f"\nCo没t total: {int(cout_total)}")


offre = [12, 11, 14, 8]
demande = [10, 11, 15, 5, 4]

couts = np.array([
    [7, 12, 1, 5, 9],
    [15, 3, 12, 6, 14],
    [8, 16, 10, 12, 7],
    [18, 8, 17, 11, 16]
])

allocation = coin_nord_ouest(offre.copy(), demande.copy())
afficher_solution(allocation, couts)
```

```python
import numpy as np

def balas_hammer(offre, demande, couts):
    m = len(offre)
    n = len(demande)
    allocation = np.zeros((m, n))

    lignes_satisfaites = [False] * m
    colonnes_satisfaites = [False] * n

    while sum(offre) > 0 and sum(demande) > 0:
        diff_lignes = []
        diff_colonnes = []

        for i in range(m):
            if not lignes_satisfaites[i]:
                couts_ligne = []
                for j in range(n):
                    if not colonnes_satisfaites[j]:
                        couts_ligne.append(couts[i, j])

                if len(couts_ligne) >= 2:
                    couts_ligne.sort()
                    diff = couts_ligne[1] - couts_ligne[0]
                else:
                    diff = couts_ligne[0] if len(couts_ligne) == 1 else 0

                diff_lignes.append((diff, i, 'ligne'))
            else:
                diff_lignes.append((-1, i, 'ligne'))

        for j in range(n):
            if not colonnes_satisfaites[j]:
                couts_colonne = []
                for i in range(m):
                    if not lignes_satisfaites[i]:
                        couts_colonne.append(couts[i, j])

                if len(couts_colonne) >= 2:
                    couts_colonne.sort()
                    diff = couts_colonne[1] - couts_colonne[0]
                else:
                    diff = couts_colonne[0] if len(couts_colonne) == 1 else 0

                diff_colonnes.append((diff, j, 'colonne'))
            else:
                diff_colonnes.append((-1, j, 'colonne'))

        max_diff_ligne = max(diff_lignes, key=lambda x: x[0])
        max_diff_colonne = max(diff_colonnes, key=lambda x: x[0])

        if max_diff_ligne[0] >= max_diff_colonne[0]:
            i = max_diff_ligne[1]
            min_cout = float('inf')
            j_min = -1
            for j in range(n):
                if not colonnes_satisfaites[j] and couts[i, j] < min_cout:
                    min_cout = couts[i, j]
                    j_min = j
            j = j_min
        else:
            j = max_diff_colonne[1]
            min_cout = float('inf')
            i_min = -1
            for i in range(m):
                if not lignes_satisfaites[i] and couts[i, j] < min_cout:
                    min_cout = couts[i, j]
                    i_min = i
            i = i_min

        alloc = min(offre[i], demande[j])
        allocation[i, j] = alloc
        offre[i] = offre[i] - alloc
        demande[j] = demande[j] - alloc

        if offre[i] == 0:
            lignes_satisfaites[i] = True
        if demande[j] == 0:
            colonnes_satisfaites[j] = True

    return allocation


def calculer_cout_total(allocation, couts):
    return np.sum(allocation * couts)


def afficher_solution(allocation, couts):
    m, n = allocation.shape

    print("\nSolution:")
    for i in range(m):
        for j in range(n):
            if allocation[i, j] > 0:
                print(f"x[{i+1},{j+1}] = {int(allocation[i, j])}")

    cout_total = calculer_cout_total(allocation, couts)
    print(f"\nCo没t total: {int(cout_total)}")


offre = [20, 17, 13]
demande = [12, 10, 15, 13]

couts = np.array([
    [3, 6, 4, 8],
    [3, 4, 7, 9],
    [9, 4, 5, 6]
])

allocation = balas_hammer(offre.copy(), demande.copy(), couts)
afficher_solution(allocation, couts)
```
